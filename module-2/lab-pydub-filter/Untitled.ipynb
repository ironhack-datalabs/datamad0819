```python
import pandas as pd
import matplotlib.pyplot as plt
from pydub import AudioSegment
from scipy import signal
import array
```

```python
audio = AudioSegment.from_ogg("PTT-20190927-WA0004.ogg")
audio
```

```python
dataaudio = audio.get_array_of_samples()
plt.plot(dataaudio)
```

```python
b, a = signal.butter(3, 1/25)
dataaudiofilter = signal.filtfilt(b, a, dataaudio)
plt.plot(dataaudiofilter)
```

```python
#audio_filtered_Array = AudioSegment.from_ogg(data_audio_filtered)

#play(audio_filtered)
audio_filtered_array = array.array(audio.array_type, dataaudiofilter.astype(int))
audio_filtered = audio._spawn(audio_filtered_array)
display(audio_filtered)

#export = audio_filtered.export("audio_filtered.ogg", format="ogg")
```
