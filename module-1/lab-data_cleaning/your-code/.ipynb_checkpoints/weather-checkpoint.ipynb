## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd

df = pd.read_csv('../weather-raw.csv')
```

```python
missing_values_count = df.isnull().sum()
print(missing_values_count)
```

```python
wdf = df.drop(columns=["id","year"])
wdf      
```

```python
tmax = wdf[wdf["element"] == "tmax"]
tmin = wdf[wdf["element"] == "tmin"]
tmax
tmin
```

```python
tmax["meanMa"]= tmax.loc[:, tmax.columns != 'month'].mean(axis=1,skipna=True)
tmax["minMa"]= tmax.loc[:, tmax.columns != 'month'].min(axis=1,skipna=True)
tmax["maxMa"]= tmax.loc[:, tmax.columns != 'month'].max(axis=1,skipna=True)
tmin["meanM"]= tmin.loc[:, tmin.columns != 'month'].mean(axis=1,skipna=True)
tmin["minM"]= tmin.loc[:, tmin.columns != 'month'].min(axis=1,skipna=True)
tmin["maxM"]= tmin.loc[:, tmin.columns != 'month'].max(axis=1,skipna=True)
```

```python
tmax = tmax[['month','meanMa','minMa','maxMa']]
tmin = tmin[['month','meanM','minM','maxM']]
```

```python
el_tiempo = pd.merge(tmax,tmin, on="month")
```
