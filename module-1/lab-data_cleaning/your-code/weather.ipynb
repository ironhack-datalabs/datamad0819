## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
```

```python
# READ FILE TO PANDAS DATAFRAME
weather = pd.read_csv('../weather-raw.csv')
display(weather)
```

```python
print(set(weather['id']),set(weather['year']))
```

```python
# DROP COLUMNS WITH CONSTANT VALUE
temp = weather.drop(columns=['id','year'])
```

```python
temp.head()
```

```python
# REMANE d1:d31 AS INTEGERS
col = list(enumerate(temp.columns[2:]))
col = {co[1]:co[0]+1 for co in col}
temp = temp.rename(columns=col)
col = list(col.values())
print(col)
```

```python
# RESHAPE TABLE WITH DAYS AS ATTRIBUTE 
temp=temp.melt(id_vars=['month','element'], value_vars=col)
temp.head()
```

```python
# DROP NaN VALUES
temp = temp.dropna()
temp.head()
```

```python
# RESHAPE TABLE WITH tmax AND tmin AS COLUMNS
temp = temp.pivot_table(values='value', index=['month','variable'], columns='element', aggfunc='first')
temp = temp.sort_index()
display(temp)
```

```python
# RENAME INDEXES AND FINAL TABLE WITH CLEANED DATA
temp.columns = pd.Index(['Max', 'Min'], name='Temperature')
temp.index.rename(['month','day'],inplace=True)
temp
```
