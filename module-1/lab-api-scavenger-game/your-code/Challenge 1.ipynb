```python
import requests
import os
from dotenv import load_dotenv
```

```python
load_dotenv(dotenv_path='.env')
github_token = os.getenv('github_token', 'No value found')
```

```python
def authRequest(url, params={}):
   headers = {
      "Authorization": "token {}".format(github_token)
   }
   response = requests.get(url,headers=headers, params=params)
   print(response.status_code)
   #print(response.headers)
   return response.json()

#data = authRequest("https://api.github.com/user/repos?type=private")
data = authRequest("https://api.github.com/repos/ironhack-datalabs/madrid-oct-2018/forks",{"sort":"newest"})
print(data)

```
