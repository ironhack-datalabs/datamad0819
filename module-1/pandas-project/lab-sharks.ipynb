```python
df = pd.read_csv(('./input/GSAF5.csv'), encoding = 'ISO-8859-1')
df.columns
```

```python
clean_df = df.drop(['pdf','href formula', 'href'], axis = 1)
clean_df
```

```python
#selecciono los datos que quiero utilizar. En este caso, quiero conocer si hace aÃ±os los tiburones atacaban mÃ¡s a las personas que acualmente
interest_data = clean_df[['Year','Date', 'Type', 'Injury','Fatal (Y/N)']]
interest_data
```

```python
#querÃ­a valorar si los tiburones matan a mÃ¡s gente pero los datos estÃ¡n incompletos, asÃ­ que tendremos que centrarnos en si los tiburones atacan mÃ¡s
interest_data[~interest_data['Fatal (Y/N)'].isin(['Y', 'N'])]

#limpieza de "fatal"
fatal = "Fatal (Y/N)"
cleanFatal = interest_data[fatal]
cleanFatal.value_counts()

#extraemos la primera conclusiÃ³n: la mayorÃ­a de los ataques de tiburones no son fatales
#para mejorar la base de datos, unificaremos los valores

interest_data['Fatal (Y/N)'] = interest_data['Fatal (Y/N)'].str.replace(' ', '').str.replace('F','Y').str.replace('#VALUE!','UNKNOWN').str.replace('n','N')
interest_data['Fatal (Y/N)'].value_counts()
```
