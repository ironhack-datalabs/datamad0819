```python
dataset = pd.read_csv("./input/GSAF5.csv")
```

```python
print("En primer lugar miro las columnas e investigo sobre ellas.")

col_dataset = dataset.columns


display(dataset[col_dataset[0:10]].head(15))
display(dataset[col_dataset[10:20]].head(15))

display(dataset[col_dataset[20:-1]].head(15))



```

```python
print("Elimino datos duplicados(si los hay) y miro a ver donde hay valores nulos")
print(dataset.shape)
dataset = dataset.drop_duplicates()
print(dataset.shape)

display(dataset.isnull().sum())
```
