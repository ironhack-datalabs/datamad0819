```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("./input/GSAF5.csv")
```

```python
print("En primer lugar miro las columnas e investigo sobre ellas.")

col_dataset = dataset.columns

display(dataset[col_dataset[0:10]].head())
display(dataset[col_dataset[10:20]].head())
display(dataset[col_dataset[20:-1]].head())
```

```python
print("Elimino datos duplicados(si los hay) y miro a ver donde hay valores nulos")
print(dataset.shape)
dataset = dataset.drop_duplicates()
print(dataset.shape)
display(dataset.isnull().sum())


```

```python
print("En un primer vistazo, las columnas 'Unnamed: 22'  y 'Unnamed: 23' no me aportan nada \
ya que el dataset tiene un total dee 5992 filas y eesas columnas tiene valores nulos de 5990\
asi que procedo a eliminarlas. Del mismo modo las columnas 'original order', 'href','href_formula'\
'Case Number.1 y Case Number.2'")

dataset = dataset.drop("Unnamed: 22",axis=1).copy()
dataset = dataset.drop("Unnamed: 23",axis=1).copy()
dataset = dataset.drop("original order",axis=1).copy()
dataset = dataset.drop("href formula",axis=1).copy()
dataset = dataset.drop("href",axis=1).copy()
dataset = dataset.drop("pdf",axis=1).copy()
dataset = dataset.drop("Case Number.1",axis=1).copy()
dataset = dataset.drop("Case Number.2",axis=1).copy()
```

```python
display(dataset.isnull().sum())

print("En las columnas con muchos valores Nan los pongo como UNKNOWN")


dataset.update(dataset[["Age"]].fillna("UNKNOWN"))
dataset.update(dataset[["Time"]].fillna("UNKNOWN"))
dataset.update(dataset[["Species "]].fillna("UNKNOWN"))

display(dataset.isnull().sum())
```

```python
print("A continuación en especies las catalogo con los mas comunes. Porque de nada me sirve tener 5000 etiquetas distintas.")


#Tiger  White  Mako  Bull Tiger Raggedtooth Hammerhead Nurse Lemon Blacktip Sevengill

from funciones import catalogarSpecies


dataset["Species "] = dataset["Species "].apply(catalogarSpecies)


```

```python
print("Las siguientes tablas a tratar son Name y Sex.")


cols_nan_sex = list(dataset['Sex '][dataset['Sex '].isnull()].index)
dataset_nan_sex = dataset.iloc[cols_nan_sex]

display(dataset_nan_sex[["Name","Sex "]].head())

print("En algunas celdas, los valores están en Nan, ya que el ataque ha sido producido a varias personas\
para ello, si en la columna Name aparece el caracter '&', pondre un codigo a Sex de 'people'")

from funciones import verNombres

#actualizo los nuevos valores en dataseet
dataset.loc[dataset_nan_sex[["Sex "]].index,"Sex "] = dataset_nan_sex[
    ["Name","Sex "]].apply(verNombres,axis = 1)

display(dataset.isnull().sum())

print("Se puede ver que se ha reducido en 200 numero de Nan, los demas valores los pondré como UNKNOWN\
Al mismo tiempo borrare la columna 'Name', ya que solo me servia por si podía rellenar mas de la columna 'Sex'")

dataset.update(dataset[["Sex "]].fillna("UNKNOWN"))
dataset = dataset.drop("Name",axis=1).copy()

print("Miro tambien si hay valores que no deberian de estar por equivocación")

display(set(dataset["Sex "]))

dataset = cleanColumn('M ','M',dataset,'Sex ')
dataset = cleanColumn('N','M',dataset,'Sex ')
dataset = cleanColumn('.','UNKNOWN',dataset,'Sex ')
dataset = cleanColumn('lli','UNKNOWN',dataset,'Sex ')


display(set(dataset["Sex "]))


```
